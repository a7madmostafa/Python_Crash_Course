### Multiple Branches


```python
num1 = float(input('Enter a number: '))
num2 = float(input('Enter anothre number: '))

if num1 > num2:
    print('The first number is greater than the second number')
elif num1 < num2:
    print('The first number is smaller than the second number')
else:
    print('The first number is equal to the second number')
```

### Simple Calculator

```python
num1 = float(input('Enter a number: '))
num2 = float(input('Enter anothre number: '))
operator= input('Enter the required operation:\n"+" for adding,"-" for subtracting, "*" for multiplication, "/" for dividing: ' )

if operator =='+':
    print('{} {} {} = {}'.format(num1,operator,num2, num1+num2))
elif operator =='-':
    print('{} {} {} = {}'.format(num1,operator,num2, num1-num2))
elif operator =='*':
    print('{} {} {} = {}'.format(num1,operator,num2, num1*num2))
elif operator =='/':
    print('{} {} {} = {}'.format(num1,operator,num2, num1/num2))
else:
    print('False operation')

```

### Nested if

```python
num = int(input('Enter an integer number: '))
if num >= 10:
   if num <=100:
        print("Between 10 and 100")
   else:
        print('More than 100')
else:
    print('Less than 10')

print('Done')
        
```

```python
num = int(input('Enter an integer number: '))
if num >= 10 and num <=100:
    print("Between 10 and 100")
elif num > 100:
    print('More than 100')
else:
    print('Less than 10')

print('Done')
```

### (2) Iterative Statements 

> __*Repeat*__ an __*action*__ based on a __*check*__

### While Loop

```python
# print Numbers from 1 to 10
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
print(7)
print(8)
print(9)
print(10)
```

```python
num =1    # counter
while num <=10:
    print(num)
    num +=1   # very important 
print('Done')    
```

### break
> Breaks out of the current closest enclosing loop.


```python
num =1    
while num <=10:
    if num == 6:
        break
    print(num)
    num +=1    
print('Done') 
```

### continue
> Goes to the top of the closest enclosing loop

```python
num =1    
while num <=10:
    if num == 6:
        num+=1
        continue
    print(num)
    num +=1    
print('Done') 
```

### pass
>  Does nothing at all

```python
num =1    
while num <=10:
    if num == 6:
        pass
    print(num)
    num +=1    
print('Done') 
```

```python
words = ["hello", "hi", "good morning", "good night"]
counter = 0
max_index = len(words) -1 

while counter <= max_index:
   word = words[counter]
   print(word + "!")
   counter = counter + 1 
```

### for loop

```python
words = ["hello", "hi", "good morning", "good night"] 

for word in words:
   print(word + "!")
```

### range
>range(start,stop,step)
* default values are start = 0 and step = 1 
* loop until value is stop - 1

```python
# print Numbers from 1 to 10
for num in range (1,11):
    print(num)
```

```python
for num in range (1,11):
    if num == 6:
        continue
    print(num)
```

```python
for num in range (1,11):
    if num == 6:
        break
    print(num)
```

```python
# Convert Celsius to Fahrenheit
celsius = range(10)

fahrenheit = [((9/5)*temp + 32) for temp in celsius ]

fahrenheit
```

### zip
>* __Syntax__ : zip (*iterators)
* __Return Value__ :Returns a single iterator object, having mapped values from all the containers.

```python
number_list = [1, 2, 3]
str_list = ['one', 'two', 'three']
zip(number_list, str_list)
```

```python
combined_list= list(zip(number_list, str_list))
print(combined_list)
```

```python
combined_set= set(zip(number_list, str_list))
print(combined_set)
```

```python
for i in zip(number_list, str_list):
    print(i)
```
